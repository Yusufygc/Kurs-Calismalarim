```python
import pandas as pd 
import numpy as np
```

### CONCAT

```python
sozluk1 = {
    "Isim": ["Ali", "Ayşe", "Fatma", "Mehmet"],
    "Spor": ["Kosu", "Yuzme", "Kosu", "Basketbol"],
    "Kalori": [100, 200, 300, 400]
   }
```

```python
df1 = pd.DataFrame(sozluk1, index=[0, 1, 2, 3])
df1
```

```python
sozluk2 = {
    "Isim": ["Ahmet", "Zeynep", "Emre", "Elif"],
    "Spor": ["Yuzme", "Basketbol", "Kosu", "Yuzme"],
    "Kalori": [150, 250, 350, 450]}
```

```python
df2 = pd.DataFrame(sozluk2, index=[4, 5, 6, 7])
df2
```

```python
sozluk3 = {
    "Isim": ["Ayse", "Mahmut", "Duygu", "Nur"],
    "Spor": ["Kosu", "Yuzme", "Badminton", "Tenis"],
    "Kalori": [300, 400,500 , 250]}
```

```python
df3 = pd.DataFrame(sozluk3, index=[8, 9, 10, 11])
df3
```

```python
pd.concat([df1, df2, df3], axis=0) # axis=0 satır bazında birleştirir
```

```python
pd.concat([df1, df2, df3], axis=1) # axis=1 sütun bazında birleştirir
#pd.concat([df1, df2, df3], axis=1, ignore_index=True) # axis=1 sütun bazında birleştirir, indexleri yok sayar
#pd.concat([df1, df2, df3], axis=1, ignore_index=True, keys=["df1", "df2", "df3"]) # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler
#pd.concat([df1, df2, df3], axis=1, ignore_index=True, keys=["df1", "df2", "df3"], sort=False) # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler, sıralama yapmaz
#pd.concat([df1, df2, df3], axis=1, ignore_index=True, keys=["df1", "df2", "df3"], sort=False, join="inner") # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler, sıralama yapmaz ve sadece ortak sütunları alır
#pd.concat([df1, df2, df3], axis=1, ignore_index=True    , keys=["df1", "df2", "df3"], sort=False, join="outer") # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler, sıralama yapmaz ve tüm sütunları alır
#pd.concat([df1, df2, df3], axis=1, ignore_index=True, keys=["df1", "df2", "df3"], sort=False, join="outer", verify_integrity=True) # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler, sıralama yapmaz ve tüm sütunları alır, veri bütünlüğünü kontrol eder
#pd.concat([df1, df2, df3], axis=1, ignore_index=True    , keys=["df1", "df2", "df3"], sort=False, join="outer", verify_integrity=True, copy=True) # axis=1 sütun bazında birleştirir, indexleri yok sayar ve anahtar isimleri ekler, sıralama yapmaz ve tüm sütunları alır, veri bütünlüğünü kontrol eder ve kopyalar
```

### MERGE

```python
mergesozluk1 = {
    "Isim": ["Ayse", "Mahmut", "Duygu", "Nur"],
    "Spor": ["Kosu", "Yuzme", "Badminton", "Tenis"],
    }
```

```python
mergedf=pd.DataFrame(mergesozluk1)
mergedf
```

```python
mergesozluk2 = {
    "Isim": ["Ayse", "Mahmut", "Duygu", "Nur"],
    "kalori": [300, 400, 500, 250]
    }
```

```python
mergedf1=pd.DataFrame(mergesozluk2)
mergedf1
```

```python
pd.merge(mergedf, mergedf1, on="Isim")# on parametresi ile ortak sütunları birleştirir
```
