```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Jupyter Notebook içinde grafiklerin otomatik olarak gösterilmesini sağlar.
```

```python
yasListesi = [10, 20, 30,30,30, 40, 50, 60, 70, 80]
kiloListesi = [50, 60, 70, 80, 90, 100, 110, 120, 130, 140]
```

```python
npYasListesi = np.array(yasListesi)
npKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(npYasListesi, npKiloListesi, color="red", marker="o", linestyle="--", linewidth=2, markersize=10) 
#color= "red" # çizgi rengini kırmızı yapar 
# marker="o" # veri noktalarını daire şeklinde gösterir
# linestyle="--" # çizgi stilini kesikli yapar
# linewidth=2 # çizgi kalınlığını 2 birim yapar
# markersize=10 # veri noktalarının boyutunu 10 birim yapar
```

```python
plt.plot(npYasListesi, npKiloListesi, color="blue", marker="o", linestyle="--", linewidth=2, markersize=10) 
plt.xlabel("Yaş") # x eksenine etiket ekler
plt.ylabel("Kilo") # y eksenine etiket ekler
plt.title("Yaş ve Kilo İlişkisi") # grafiğe başlık ekler
```

```python
numpyDizisi1 =np.linspace(0, 10, 20) # 0 ile 10 arasında 100 eşit parçaya bölünmüş değerler oluşturur
numpyDizisi1
```

```python
numpyDizisi2  = numpyDizisi1 ** 2 # numpyDizisi1'in her elemanını 2 us alir
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--") # "g--" ifadesi, yeşil renkte kesikli bir çizgi çizer \ g*-
```

```python
plt.subplot(1,2,1) # 1 satır, 2 sütunluk bir alt grafik düzeni oluşturur ve ilk alt grafiği seçer
plt.plot(numpyDizisi1, numpyDizisi2, "g--") # İlk alt grafikte yeşil kesikli çizgi çizer
plt.title("Alt Grafik 1") # İlk alt grafiğe başlık ekler
plt.subplot(1,2,2) # 1 satır, 2 sütunluk bir alt grafik düzeni oluşturur ve ikinci alt grafiği seçer
plt.plot(numpyDizisi2, numpyDizisi1, "r*-")
# İkinci alt grafikte kırmızı yıldız işaretli çizgi çizer
plt.title("Alt Grafik 2") # İkinci alt grafiğe başlık ekler
```

🧩 `plt.figure()` Nedir?

- Yeni bir çizim alanı (figure) başlatır.
- İçine bir veya daha fazla alt grafik (subplot/axes) yerleştirilebilir.
- Özellikle birden fazla grafik çizeceğiniz zaman veya grafiklerin boyutlarını ve özelliklerini özelleştirmek istediğinizde kullanılır.


```python
benimFigure = plt.figure() # Yeni bir figür oluşturur
figureaxes =benimFigure.add_axes([0.2,0.2,0.9,0.9]) # Eksen ekler, [sol, alt, genişlik, yükseklik] şeklinde normalleştirilmiş koordinatlar kullanır
figureaxes.plot(numpyDizisi1, numpyDizisi2, "b--") # Eksen üzerinde mavi kesikli çizgi çizer
figureaxes.set_xlabel("X Ekseni") # X eksenine etiket ekler
figureaxes.set_ylabel("Y Ekseni") # Y eksenine etiket ekler
figureaxes.set_title("Özel Eksen Grafiği") # Eksen başlığı ekler
```
