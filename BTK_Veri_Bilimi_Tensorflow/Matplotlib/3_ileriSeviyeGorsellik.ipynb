```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 =np.linspace(0, 10, 20)
```

```python
numpyDizisi2  = numpyDizisi1 ** 2 
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color ="#3A95A8", alpha = 0.3)
benimEksen.plot(numpyDizisi2, numpyDizisi1, color ="#C96F23")
```

```python
(yeniFigur, yeniEksen) = plt.subplots() # Figure boyutunu ayarlar
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +2 , color ="blue", linewidth = 2, linestyle = "--", label = "numpy dizisi +2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +3 , color ="red", linewidth = 2, linestyle = "-.", label = "numpy dizisi +3")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +4 , color ="yellow", linewidth = 3, linestyle = "-", label = "numpy dizisi +4")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +5 , color ="green", linewidth = 3, linestyle = ":", label = "numpy dizisi +5")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +6 , color ="#000000", linewidth = 3, linestyle = "--",marker="o",markersize = 5,markerfacecolor = "red", label = "numpy dizisi +6")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 +7 , color ="#000000", linewidth = 1, linestyle = "--",marker="+",markersize = 9, label = "numpy dizisi +6")



yeniEksen.legend()
```
