```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 =np.linspace(0, 10, 20)
```

```python
numpyDizisi2  = numpyDizisi1 ** 2 
```

## SCATTER

```python
plt.scatter(numpyDizisi1, numpyDizisi2, color ="#3A95A8")
```

## HISTOGRAM

```python
yeniDizi = np.random.randint(0,100,50)
yeniDizi
```

```python
plt.hist(yeniDizi, bins=10, color="#C96F23")
```

## BOXPLOT

```python
plt.boxplot(yeniDizi, vert=True, patch_artist=True, boxprops=dict(facecolor="#3A95A8", color="#C96F23"), medianprops=dict(color="black"))
```

📌 Parametre Açıklamaları:

### 1. `yeniDizi`
- Görselleştirilecek veridir. Tek boyutlu veya çok boyutlu bir liste/array olabilir.
- Eğer çok boyutlu bir array verilirse, her bir iç liste ayrı bir boxplot olarak gösterilir.

---

### 2. `vert=True`
- Kutu grafiğinin dikey mi yoksa yatay mı olacağını belirler.
    - **True** ➝ Dikey kutu grafiği.
    - **False** ➝ Yatay kutu grafiği.

---

### 3. `patch_artist=True`
- Kutu kutularının dolgu (renk) ile boyanmasını sağlar.
    - Eğer **True** olmazsa, `boxprops` içindeki `facecolor` uygulanmaz.
- Dolgu rengini etkinleştirir ve kutuların görsel olarak daha belirgin olmasını sağlar.

---

### 4. `boxprops=dict(facecolor="#3A95A8", color="#C96F23")`
- Kutu kutusunun stilini tanımlar.
    - **`facecolor="#3A95A8"`** ➝ Kutunun iç dolgu rengi (mavi-yeşil ton).
    - **`color="#C96F23"`** ➝ Kutunun kenarlık rengi (turuncu ton).
- **Not:** `patch_artist=True` olmalı ki `facecolor` çalışsın.

---

### 5. `medianprops=dict(color="black")`
- Ortanca (median) çizgisinin stilini tanımlar.
    - **`color="black"`** ➝ Ortanca çizgisi siyah olacak.

---

Bu parametreler, kutu grafiğinin görünümünü ve davranışını özelleştirmek için kullanılır. Örneğin, `yeniDizi` verisi görselleştirilirken, kutuların dolgu rengi, kenarlık rengi ve ortanca çizgisi gibi detaylar bu parametrelerle kontrol edilir.
