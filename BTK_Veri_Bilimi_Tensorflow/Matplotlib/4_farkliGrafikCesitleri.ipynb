```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 =np.linspace(0, 10, 20)
```

```python
numpyDizisi2  = numpyDizisi1 ** 2 
```

## SCATTER

```python
plt.scatter(numpyDizisi1, numpyDizisi2, color ="#3A95A8")
```

## HISTOGRAM

```python
yeniDizi = np.random.randint(0,100,50)
yeniDizi
```

```python
plt.hist(yeniDizi, bins=10, color="#C96F23")
```

## BOXPLOT

```python
plt.boxplot(yeniDizi, vert=True, patch_artist=True, boxprops=dict(facecolor="#3A95A8", color="#C96F23"), medianprops=dict(color="black"))
```

üìå Parametre A√ßƒ±klamalarƒ±:

### 1. `yeniDizi`
- G√∂rselle≈ütirilecek veridir. Tek boyutlu veya √ßok boyutlu bir liste/array olabilir.
- Eƒüer √ßok boyutlu bir array verilirse, her bir i√ß liste ayrƒ± bir boxplot olarak g√∂sterilir.

---

### 2. `vert=True`
- Kutu grafiƒüinin dikey mi yoksa yatay mƒ± olacaƒüƒ±nƒ± belirler.
    - **True** ‚ûù Dikey kutu grafiƒüi.
    - **False** ‚ûù Yatay kutu grafiƒüi.

---

### 3. `patch_artist=True`
- Kutu kutularƒ±nƒ±n dolgu (renk) ile boyanmasƒ±nƒ± saƒülar.
    - Eƒüer **True** olmazsa, `boxprops` i√ßindeki `facecolor` uygulanmaz.
- Dolgu rengini etkinle≈ütirir ve kutularƒ±n g√∂rsel olarak daha belirgin olmasƒ±nƒ± saƒülar.

---

### 4. `boxprops=dict(facecolor="#3A95A8", color="#C96F23")`
- Kutu kutusunun stilini tanƒ±mlar.
    - **`facecolor="#3A95A8"`** ‚ûù Kutunun i√ß dolgu rengi (mavi-ye≈üil ton).
    - **`color="#C96F23"`** ‚ûù Kutunun kenarlƒ±k rengi (turuncu ton).
- **Not:** `patch_artist=True` olmalƒ± ki `facecolor` √ßalƒ±≈üsƒ±n.

---

### 5. `medianprops=dict(color="black")`
- Ortanca (median) √ßizgisinin stilini tanƒ±mlar.
    - **`color="black"`** ‚ûù Ortanca √ßizgisi siyah olacak.

---

Bu parametreler, kutu grafiƒüinin g√∂r√ºn√ºm√ºn√º ve davranƒ±≈üƒ±nƒ± √∂zelle≈ütirmek i√ßin kullanƒ±lƒ±r. √ñrneƒüin, `yeniDizi` verisi g√∂rselle≈ütirilirken, kutularƒ±n dolgu rengi, kenarlƒ±k rengi ve ortanca √ßizgisi gibi detaylar bu parametrelerle kontrol edilir.
