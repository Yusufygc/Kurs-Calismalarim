```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 =np.linspace(0, 10, 20)
```

```python
numpyDizisi2  = numpyDizisi1 ** 2 
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.22,0.42,0.3,0.3]) # Kucuk Grafik ilk iki parametre grafigin konumunu ayarlar

eksen1.plot(numpyDizisi1, numpyDizisi2, 'g')
eksen1.set_xlabel('X Ekseni')
eksen1.set_ylabel('Y Ekseni')
eksen1.set_title('Ana Grafik Baslik')

eksen2.plot(numpyDizisi2, numpyDizisi1, 'b')
eksen2.set_xlabel('Kucuk Grafik X Ekseni')
eksen2.set_ylabel('Kucuk Grafik Y Ekseni')
eksen2.set_title('Kucuk Grafik Baslik')
```

```python
plt.subplot() # eksen de geliyor
```

```python
plt.figure() # eksen yok bunda
```

```python
(benimFigure, benimEksen) = plt.subplots() # eksen de geliyor
benimEksen.plot(numpyDizisi1, numpyDizisi2, 'r')
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2)
type(benimEksen) # Eksenler 2D dizisi olarak geliyor

for eksen in benimEksen:
    eksen.plot(numpyDizisi1, numpyDizisi2, 'r')
    eksen.set_xlabel('X Ekseni')
plt.tight_layout() # Alt alta gelmesini engeller daha iyi gorunmesini saglar
```
